**문제 5: 조건부 선택 (복합 조건)**

1. 남성이면서 1등급인 승객들의 수를 구하세요.

조건이 여러개 일때 각 조건을 괄호로 반드시 감싸고 조건을 의미에 맞게 `&`, `|`로 적당히 연결하면 됩니다.



```python
# 여기 코드를 입력하세요.
print(sum((titanic['Sex'] == 'male') & (titanic['Pclass'] == 1)))
```

2. 여성이면서 생존한 승객들 중 첫 3명의 'Name', 'Age', 'Pclass' 정보를 출력하세요.

아래 코드처럼 코드가 한줄에 너무 길게 진행되면 (), [] 를 기준으로 적당히 줄바꿈해서 보기 쉽게하는 것도 답안을 작성하는 요령입니다.


```python
# 여기 코드를 입력하세요.
print(
    titanic[
        (titanic['Sex'] == 'female') & (titanic['Survived'] == 1)
    ].iloc[:3][['Name', 'Age', 'Pclass']]
)
```

3. 나이가 10세 미만이거나 60세 이상인 승객들의 수를 구하세요.

4. Pclass가 2등급 또는 3등급이면서 생존한 승객들 중 'Name', 'Pclass', 'Survived' 정보를 첫 5명만 출력하세요.

이 문제는 `&`와 `|`가 함께 사용되는 문제입니다. 조건이 좀 복잡하기 때문에 천천히 생각하면서 코드로 바꾸는 연습을 충분히 하도록 하세요.



```python
# 여기 코드를 입력하세요.
print(
    titanic[
        ((titanic['Pclass'] == 2) | (titanic['Pclass'] == 3))
        &
        (titanic['Survived'] == 1)
    ][['Name', 'Pclass', 'Survived']].iloc[:5]
)
```

**문제 6: 종합 문제**
다음 분석을 수행하세요.
1. 각 등급(Pclass)별 여성 승객의 생존율을 비교하기 위해, 1등급 여성 승객 중 생존자 수와 전체 1등급 여성 승객 수를 각각 구하여 생존률을 계산하세요.





```python
# 여기 코드를 입력하세요.
titanic_female_pclass1 = titanic[(titanic['Pclass'] == 1) & (titanic['Sex'] == 'female')]
survival_rate = titanic_female_pclass1['Survived'].sum() / titanic_female_pclass1.shape[0]

print(f"{survival_rate:.2%}")
```

분석 결과 1등급 여성 승객들은 대부분 생존했음을 알 수 있습니다.

2. 각 등급(Pclass)별 생존율을 계산하여 비교하세요. 1등급, 2등급, 3등급 각각의 (생존자 수 / 전체 승객 수)를 구하고, 어느 등급의 생존율이 가장 높은지 분석하세요.





```python
# 여기 코드를 입력하세요.
# 1등급
pclass1 = titanic[titanic['Pclass'] == 1]
survival_rate_1 = pclass1['Survived'].sum() / pclass1.shape[0]
print(f"1등급 생존율: {survival_rate_1:.2%} ({pclass1['Survived'].sum()}/{pclass1.shape[0]})")

# 2등급
pclass2 = titanic[titanic['Pclass'] == 2]
survival_rate_2 = pclass2['Survived'].sum() / pclass2.shape[0]
print(f"2등급 생존율: {survival_rate_2:.2%} ({pclass2['Survived'].sum()}/{pclass2.shape[0]})")

# 3등급
pclass3 = titanic[titanic['Pclass'] == 3]
survival_rate_3 = pclass3['Survived'].sum() / pclass3.shape[0]
print(f"3등급 생존율: {survival_rate_3:.2%} ({pclass3['Survived'].sum()}/{pclass3.shape[0]})")
```

결과를 보면 높은 등급의 객실을 이용한 승객들이 더 많이 생존했다는 사실을 명확히 확인할 수 있습니다.

3. 여성 승객과 남성 승객의 생존율을 각각 구하고, 18세 미만 어린이와 성인 남성(18세 이상 남성)의 생존율을 각각 구하세요. "여성과 어린이 먼저" 원칙이 실제로 적용되었는지 분석하세요.





```python
# 여기 코드를 입력하세요.
# 여성 생존율
female = titanic[titanic['Sex'] == 'female']
female_survival = female['Survived'].sum() / female.shape[0]
print(f"여성 생존율: {female_survival:.2%} ({female['Survived'].sum()}/{female.shape[0]})")

# 남성 생존율
male = titanic[titanic['Sex'] == 'male']
male_survival = male['Survived'].sum() / male.shape[0]
print(f"남성 생존율: {male_survival:.2%} ({male['Survived'].sum()}/{male.shape[0]})")

# 어린이(18세 미만) 생존율
children = titanic[titanic['Age'] < 18]
children_survival = children['Survived'].sum() / children.shape[0]
print(f"\n어린이(18세 미만) 생존율: {children_survival:.2%} ({children['Survived'].sum()}/{children.shape[0]})")

# 성인 남성(18세 이상 남성) 생존율
adult_male = titanic[(titanic['Age'] >= 18) & (titanic['Sex'] == 'male')]
adult_male_survival = adult_male['Survived'].sum() / adult_male.shape[0]
print(f"성인 남성(18세 이상) 생존율: {adult_male_survival:.2%} ({adult_male['Survived'].sum()}/{adult_male.shape[0]})")
```

여성(74.2%)과 어린이(54.0%)의 생존율이 성인 남성(17.7%)보다 높아, '여성과 어린이 먼저' 원칙이 실제로 적용되었음을 알 수 있습니다.
