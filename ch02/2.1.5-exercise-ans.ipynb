**문제 4: 데이터 분석**

앞 문제 3-3에서 평균을 구할 때 `합 / 개수` 방식을 사용했는데, Pandas의 `DataFrame['열이름'].mean()` 함수를 사용하면 더 간단하게 평균을 계산할 수 있습니다. 이 문제부터는 `.mean()` 함수를 활용하세요.

다음 분석을 수행하세요.





1. 나이가 18세 미만인 어린이의 생존율과 18세 이상 남성 성인의 생존율을 계산하여 비교하세요.

복잡한 조건의 문제가 주어지면 모든 작업을 한줄의 코드로 작성할 수 없습니다. 이런 경우 필요한 변수를 얼마든지 만들어도 상관없기 때문에 답안처럼 작성하기를 권장합니다.

```python
# 여기 코드를 입력하세요.

```

**문제 5: 메서드 체이닝 종합**

메서드 체이닝은 여러 작업을 연속적으로 연결하여 한 번에 수행하는 방식입니다. 앞서 배운 것처럼 Pandas의 대부분 메서드는 새로운 DataFrame을 반환하므로, 이를 이용해 여러 메서드를 `.`으로 연결할 수 있습니다.

간단한 예시를 먼저 살펴보겠습니다:

```python
# 메서드 체이닝 예시: 여성 승객을 선택하고 나이순으로 정렬하여 상위 3명 출력
# 읽기 쉽게 여러 줄로 작성
print(
    titanic[titanic['Sex'] == 'female']
    .sort_values('Age')
    [['Name', 'Age']]
    .head(3)
)
```

실행 결과:
```
                              Name   Age
644         Baclini, Miss. Eugenie  0.75
469  Baclini, Miss. Helene Barbara  0.75
172   Johnson, Miss. Eleanor Ileen  1.00
```

`titanic[titanic['Sex'] == 'female']`여기까지 실행 결과가 DataFrame을 반환하면 연이어 `.sort_values('Age')`를 실행하여 다시 반환된 DataFrame에 `[['Name', 'Age']].head(3)` 적용합니다.

이제 메서드 체이닝을 활용하여 다음 작업을 수행하세요.
