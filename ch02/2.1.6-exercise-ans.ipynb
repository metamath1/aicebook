**문제: 데이터 병합과 분석**

1. 전체 주문에서 총 판매 수량(quantity 합계)이 가장 많은 아이템은 무엇인가?

order 데이터프레임에 다양한 제품들이 있기 때문에 제품 이름을 그룹화하고 'quantity'를 합산합니다.

```python
# 여기 코드를 입력하세요.
print("아이템별 총 판매 수량:")
print(
    orders.groupby('item')['quantity'].sum().
    sort_values(ascending=False)
)
```

실행 결과:

실행 결과 수건이 5개로 가장 판매량이 많은것을 알 수 있습니다.



2. '영희'가 구매한 아이템 중 총 구매 수량(quantity 합계)이 가장 많은 아이템은 무엇인가?

고객과 주문을 합쳐야 하므로 `merge()`를 사용합니다. 짝이 없는 주문이나 고객은 고려할 필요없으므로 `how='inner'`로 지정합니다.

```python
# 여기 코드를 입력하세요.
# customer와 orders를 merge
customer_orders = pd.merge(customer, orders, on='cust_id', how='inner')

# 영희의 주문만 필터링
younghee_orders = customer_orders[customer_orders['name'] == '영희']
print("영희의 주문 내역:")
print(younghee_orders)

# 영희가 구매한 아이템별 수량 합계
print("\n영희의 아이템별 구매 수량:")
print(
    younghee_orders.groupby('item')['quantity'].sum()
    .sort_values(ascending=False)
)
```

실행 결과:
```
영희의 주문 내역:
   cust_id name  나이 item  quantity
0        1   영희  20   치약         1
1        1   영희  20   칫솔         2
2        1   영희  20   치약         3

영희의 아이템별 구매 수량:
item
치약    4
칫솔    2
Name: quantity, dtype: int64
```

영희의 주문 내역을 만들고 난 다음 1번 문제처럼 제품 이름으로 그룹화하겨 수량을 모두 더하면 영희는 치약을 가장 많이 구입했다는 것을 알 수 있습니다.

3. 고객 정보가 없는 주문(cust_id가 customer에 없는 경우)의 총 주문 건수는?

고객 정보가 없는 주문이라도 없어지면 안되므로 왼쪽에 customer 오른쪽에 orders를 두고 `how='right'`로 지정하여 병합합니다. 병합된 데이터프레임에 고객정보가 없는 데이터를 `isna()`를 사용하여 필터링 합니다.

```python
# 여기 코드를 입력하세요.
# right merge로 모든 주문 데이터 유지
all_orders = pd.merge(customer, orders, on='cust_id', how='right')

# 고객 정보가 없는 주문 (name이 NaN인 경우)
no_customer_orders = all_orders[all_orders['name'].isna()]
print("고객 정보가 없는 주문:")
print(no_customer_orders)
```

실행 결과:
```
고객 정보가 없는 주문:
   cust_id name  나이 item  quantity
9        9  NaN NaN  케이스         1
```

고객 정보가 없는 주문은 한건 있습니다.

4. 한 번도 주문하지 않은 고객의 이름을 모두 출력하시오.

이번에는 왼쪽 데이터프레임의 모든 데이터를 남겨서 제대로 병합되지 않은 데이터를 필터링하는 문제로 3번 문제와 반대 경우입니다.

```python
# 여기 코드를 입력하세요.
# left merge로 모든 고객 정보 유지
all_customers = pd.merge(customer, orders, on='cust_id', how='left')

# 주문이 없는 고객 (item이 NaN인 경우)
no_order_customers = all_customers[all_customers['item'].isna()]
print("주문하지 않은 고객:")
print(no_order_customers[['cust_id', 'name', '나이']])
```

실행 결과:
```
주문하지 않은 고객:
    cust_id name  나이
0         0   철수  40
10        5   동건  18
```
