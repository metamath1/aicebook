#### 문제 12
박스플롯을 보면 80세 근처에 이상치가 존재하는 것을 확인할 수 있습니다. 이상치가 있을 때는 평균보다 중앙값으로 결측치를 채우는 것이 더 안정적입니다. 이상치가 확인되므로 `age` 변수의 결측치는 성별로 구분하여 중앙값으로 채우세요.

```python
#여기 코드를 입력하세요.
# 남성의 나이 중앙값 계산 및 결측치 채우기
med_age_male = titanic_2[titanic_2['sex'] == 'male']['age'].median()
titanic_2.loc[(titanic_2['sex'] == 'male') & titanic_2['age'].isnull(), 'age'] = med_age_male

# 여성의 나이 중앙값 계산 및 결측치 채우기
med_age_female = titanic_2[titanic_2['sex'] == 'female']['age'].median()
titanic_2.loc[(titanic_2['sex'] == 'female') & titanic_2['age'].isnull(), 'age'] = med_age_female

print(titanic_2.isnull().sum())
```

실행 결과:
```
pclass      0
survived    0
sex         0
age         0
sibsp       0
parch       0
fare        1
embarked    2
dtype: int64
```

우선 중앙값을 구해 `titanic_2['sex'] == 'male'` 조건으로 필터링해서 구해둡니다. 다음 `loc[]`을 사용합니다. 행에 남자면서 `age`가 결측치인 행을 `(titanic_2['sex'] == 'male') & titanic_2['age'].isnull()`로 지정하고 컬럼을 `age`로 두고 계산한 중앙값을 대입합니다.

시험에 나올 수 있는 사용법이니 꼭 기억해두시길 바랍니다.

#### 문제 13
`fare`, `embarked`에 결측치를 가진 데이터는 삭제합니다.

```python
#여기 코드를 입력하세요.
titanic_2 = titanic_2[(~titanic_2['fare'].isnull()) & (~titanic_2['embarked'].isnull())].copy()
print(titanic_2.isnull().sum())
```

실행 결과:
```
pclass      0
survived    0
sex         0
age         0
sibsp       0
parch       0
fare        0
embarked    0
dtype: int64
```

> **참고:** pandas에서 데이터프레임을 조작하다 보면 SettingWithCopyWarning이라는 경고가 발생하는 경우가 있습니다. 이 경고가 시험에서 큰 문제를 만들 가능성은 없지만 경고를 확실히 없애기 위해 `.copy()`를 사용하여 명확하게 복사본을 만들었습니다. 이렇게 하면 이후 작업에서 SettingWithCopyWarning이 발생하지 않습니다.

#### 문제 18
`age`, `fare` 변수를 Min-Max scaling 하세요.

```python
#여기 코드를 입력하세요.
# age 변수 Min-Max scaling
age_min = titanic_3['age'].min()
age_max = titanic_3['age'].max()
titanic_3['age'] = (titanic_3['age'] - age_min) / (age_max - age_min)

# fare 변수 Min-Max scaling
fare_min = titanic_3['fare'].min()
fare_max = titanic_3['fare'].max()
titanic_3['fare'] = (titanic_3['fare'] - fare_min) / (fare_max - fare_min)

print(titanic_3.head())
```

실행 결과:
```
   pclass  survived     sex       age      fare embarked  family  is_alone
0       1         1  female  0.361169  0.412503        S       0         1
1       1         1    male  0.009395  0.295806        S       3         0
2       1         0  female  0.022964  0.295806        S       3         0
3       1         0    male  0.373695  0.295806        S       3         0
4       1         0  female  0.311064  0.295806        S       3         0
```

2.3.4에서 배운것처럼 컬럼 단위로 직접 스케일링을 계산했습니다. 다시 언급하지만 스케일링은 scikit-learn 라이브러리를 사용하는것이 보통입니다. 하지만 데이터프레임 컬럼을 직접 다루는 연습을 스케일링을 통해 해보는 것도 나쁘지 않습니다.
