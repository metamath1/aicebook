1. pandas, matplotlib, seaborn 라이브러리를 로딩하고 `sns.load_dataset('tips')로 tips 데이터를 로드하고 데이터를 탐색하세요


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 팁 데이터 로드
tips = sns.load_dataset('tips')

print("데이터 크기:", tips.shape)
print("\n처음 5개 데이터:")
print(tips.head())

print("\n기본 통계:")
print(tips.describe())
```

2. 산점도를 그려보세요.

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
plt.plot(tips['total_bill'], tips['tip'], 'o')
plt.xlabel('total_bill')
plt.xlabel('tip')
plt.show()
```

3. total_bill을 특성으로, tip을 타겟으로 X, y에 저장하세요.


```python
# 특성(X)과 타겟(y) 분리
X = tips[['total_bill']]   # 2차원 배열로 주의하세요!
y = tips['tip']            # 1차원 배열
```

4. 회귀 모델을 생성하고 모델을 학습하고 예측하세요


```python
# 모델 생성 및 학습
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
```

5. MAE, MSE, RMSE, R² 지표로 모델을 평가하세요


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import root_mean_squared_error, r2_score


mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = root_mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"MAE: {mae:.2f}, MSE: {mse:.2f}, RMSE: {rmse:.2f}, R2: {r2:.2f}")
```
