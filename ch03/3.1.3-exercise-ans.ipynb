1. 데이터를 로드하고 탐색하세요.

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

# 데이터 로드
cancer_data = load_breast_cancer()

# DataFrame으로 변환
data = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
y = pd.Series(cancer_data.target)

print(f"데이터 크기: {data.shape}")
print(f"\n타겟 분포:")
print(f"악성(0): {(y==0).sum()}개")
print(f"양성(1): {(y==1).sum()}개")

print("특성")
data.columns
```

2. 'mean perimeter'를 입력 특성으로 하여 로지스틱 회귀 모델을 학습시키세요. (타이타닉과는 다르게 꽤 좋은 결과를 얻을 수 있습니다.)


```python
from sklearn.linear_model import LogisticRegression

X = data[['mean perimeter']]

logreg = LogisticRegression()
logreg.fit(X, y)
```

3. 혼동행렬과 classification_report를 출력하세요.


```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = logreg.predict(X)

print(confusion_matrix(y, y_pred))

print(classification_report(y, y_pred))
```

4. 정확도, 정밀도, 재현율, F1-Score를 계산하세요


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

acc = accuracy_score(y, y_pred)
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)
f1 = f1_score(y, y_pred)

print(acc, precision, recall, f1)
```
